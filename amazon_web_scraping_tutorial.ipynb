```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

headers = {"User-Agent": ""}#add your user agent 
url = "https://www.amazon.in/s?k=mobiles&crid=2F49SZUME2VJO&sprefix=mobiles%2Caps%2C292&ref=nb_sb_noss_1"
r = requests.get(url,headers=headers)
r
```

```python
# Soup Object containiang all data
soup  = BeautifulSoup(r.content,"html.parser")
soup
```

```python
#extracting data from list of items
links = soup.find_all("a",attrs={'class':"a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"})
links
```

```python
#getting address of index 0 element in links
link = links[0].get('href')
link
```

```python
#creating a single element link
product_link = "https://www.amazon.in"+link
product_link
```

```python
#accessing the link page and code
new_page = requests.get(product_link,headers=headers)
new_soup = BeautifulSoup(new_page.content,'html.parser')
new_soup
```

```python
#name of the product
y = new_soup.find('span',attrs={"id":"productTitle"}).text.strip()
y
```

```python
#price of the product
new_soup.find('span',attrs={'class':"a-price-whole"}).text
```

```python
#customer reviews of the product
new_soup.find('span',attrs={'id':"acrCustomerReviewText"}).text
```

```python
#avaliability of the product
new_soup.find('div',attrs={'id':"availability"}).find('span').text.strip()
```

```python
#rating of the product
new_soup.find('span',attrs={'data-hook':"rating-out-of-text"}).text
```
