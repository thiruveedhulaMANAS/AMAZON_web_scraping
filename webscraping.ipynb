```python
#importing python modules
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36 Edg/107.0.1418.52"}
url = "https://www.amazon.in/s?k=watches&crid=3CI6LEWRQHVXK&sprefix=watches%2Caps%2C283&ref=nb_sb_noss_1"
r = requests.get(url,headers=headers)
soup  = BeautifulSoup(r.content,"html.parser")
links = soup.find_all("a",attrs={'class':"a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"})
ls = []
for i in links:
    ls.append("https://www.amazon.in"+i.get('href'))
name=[]
price=[]
rating=[]
stock=[]
stars=[]
```

```python
def get_name(new_soup):
    try:
        n = new_soup.find('span',attrs={"id":"productTitle"}).text.strip()
    except AttributeError:
        n=''
    return n

def get_price(new_soup):
    try:
        x = new_soup.find('span',attrs={'class':"a-price-whole"}).text
        n = '₹'+x+'00'
    except AttributeError:
        n = ''
    return n

def get_rating(new_soup):
    try:
        n = new_soup.find('span',attrs={'id':"acrCustomerReviewText"}).text
    except AttributeError:
        n = ''
    return n

def get_availability(new_soup):
    try:
        n = new_soup.find('div',attrs={'id':"availability"}).find('span').text.strip()
    except AttributeError:
        n = ''
    return n

def get_stars(new_soup):
    try:
        n = new_soup.find('span',attrs={'data-hook':"rating-out-of-text"}).text
    except:
        n = ''
    return n


```

```python
for i in ls:
    new_page = requests.get(i,headers=headers)
    new_soup = BeautifulSoup(new_page.content,'html.parser')
    name.append(get_name(new_soup))
    price.append(get_price(new_soup))
    rating.append(get_rating(new_soup))
    stock.append(get_availability(new_soup))
    stars.append(get_stars(new_soup))
```

```python
#storing extracted data in csv file
data = {'Name':name,
'Price':price,
'CustomerReviewText':rating,
'Availability':stock,
'Ratings':stars}
df = pd.DataFrame(data)
df.to_csv('book1.csv',header=True,index=False)
```
